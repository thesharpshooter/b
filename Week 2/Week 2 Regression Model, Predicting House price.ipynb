```python
import graphlab as gl
# import matplotlib.pyplot as plt
gl.canvas.set_target('ipynb')
sales = gl.SFrame("home_data.gl/")
print sales["price","zipcode"]
```

```python
sales.show(view="BoxWhisker Plot", x = "zipcode", y = "price")
training_data,test_data = sales.random_split(0.8,seed=0)
```

```python
#model based on sqrtft_living as the only feature
model_1 = gl.regression.create(training_data,target="price",features=["sqft_living"])
model_1_linear = gl.linear_regression.create(training_data,target="price",features=["sqft_living"])
```

```python
print model_1.evaluate(training_data)
print model_1_linear.evaluate(training_data)

# print model_1.predict(test_data)
# print model_1_linear.predict(test_data)
```
