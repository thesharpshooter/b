# A function that returns the input matrix and output

```python
def get_numpy_data(data, features, output):
    data['ones'] = 1
    features = ['ones'] + features
    return data[features],data[output]
```

# A function that predicts output

```python
def predict_output(feature_matrix,weights):
    return np.dot(feature_matrix,weights)
```

# A function that returns the feature derivate ie. sum((y-y_hat)*h)

```python
def feature_derivative(error,feature):
    return -2*np.dot(feature.transpose(),error)
```

```python
def regression_gradiant_descent(feature_matrix,output,initial_weights,step_size,tolerance):
    weights = initial_weights
    prediction = predict_output(feature_matrix,weights)
    gradiant = feature_derivative(output-prediction,feature_matrix)
    mag = np.math.sqrt((gradiant**2).sum())
    print weights
    print prediction
    print gradiant
    print mag
    plt.plot(feature_matrix,output,'o')
    
    while mag > tolerance :
        weights = weights - step_size*gradiant
        
        prediction = predict_output(feature_matrix,weights)
        gradiant = feature_derivative(output-prediction,feature_matrix)
        mag = np.math.sqrt((gradiant**2).sum())
        print weights
        print prediction
        print gradiant
        print mag
    plt.plot(feature_matrix,output,'o',feature_matrix,prediction)
    return weights
```

```python
simple_features = ['sqft_living']
my_output= 'price'
(simple_feature_matrix, output) = get_numpy_data(data_train, simple_features, my_output)
initial_weights = np.array([-47000., 1.])
step_size = 7e-12
tolerance = 2.5e7
```

```python
weights = regression_gradiant_descent(simple_feature_matrix,output,initial_weights,step_size,tolerance)
```

```python
simple_features = ['sqft_living']
my_output= 'price'

(test_simple_feature_matrix, output) = get_numpy_data(data_test, simple_features, my_output)

prediction_test = predict_output(test_simple_feature_matrix,weights)
print prediction_test[0]
```

```python
model_features = ['sqft_living', 'sqft_living15']
my_output = 'price'
(feature_matrix, output) = get_numpy_data(data_train, model_features,my_output)
initial_weights = np.array([-100000., 1., 1.])
step_size = 4e-12
tolerance = 1e9
```

```python
weights_2 = regression_gradiant_descent(feature_matrix,output,initial_weights,step_size,tolerance)
```

```python

(test_model_feature_matrix, output) = get_numpy_data(data_test, model_features, my_output)
prediction_test_model2 = predict_output(test_model_feature_matrix,weights_2)
print prediction_test_model2[0]
```
