# Regression problem using the gradient descent method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.DataFrame(np.arange(0,5),columns = ['x'])
data['y'] = np.array([1,3,7,13,21])
data
```

```python
plt.plot(data['x'],data['y'],'o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Initializing w0, w1, step, tolerance 

```python
w0 = 0 #intercept
w1 = 0 #slope
step = 0.01
tolerance = 0.01
i = 0
```

```python
data['y_predicted'] = w0 + w1*data['x']
data['error'] = data['y'] - data['y_predicted']
data['error_x'] = data['error']*data['x']
sum_error = data['error'].sum()
sum_errorx = data['error_x'].sum()
mag = np.math.sqrt(sum_error**2 + sum_errorx**2)
print data
print "Sum error : ",sum_error
print "Sum error x : ",sum_errorx
print "Gradiant mag : ",mag
plt.plot(data['x'],data['y'],'o',data['x'],data['y_predicted'])
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
while mag > tolerance:
    w0 = w0 + 2*step*sum_error
    w1 = w1 + 2*step*sum_errorx
    i += 1
    
    print "w0 : ",w0
    print "w1 : ",w1
    
    data['y_predicted'] = w0 + w1*data['x']
    data['error'] = data['y'] - data['y_predicted']
    data['error_x'] = data['error']*data['x']
    sum_error = data['error'].sum()
    sum_errorx = data['error_x'].sum()
    mag = np.math.sqrt(sum_error**2 + sum_errorx**2)
    
    print data
    print "Sum error : ",sum_error
    print "Sum error x : ",sum_errorx
    print "Gradiant mag : ",mag
    plt.plot(data['x'],data['y'],'o',data['x'],data['y_predicted'])
    plt.xlabel('X axis')
    plt.ylabel('Y axis')

print "Final w0 : ",w0
print "Final w1 : ",w1
print "FInal total error : ",sum_error
print "TOtal no of iterations",i
```

```python
print data
print "Sum error : ",sum_error
print "Sum error x : ",sum_errorx
print "Gradiant mag : ",mag
plt.plot(data['x'],data['y'],'o',data['x'],data['y_predicted'])
plt.xlabel('X axis')
plt.ylabel('Y axis')
print "Final w0 : ",w0
print "Final w1 : ",w1
print "FInal total error : ",sum_error
print "TOtal no of iterations",i
```
